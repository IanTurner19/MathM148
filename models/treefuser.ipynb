```python
import matplotlib.pyplot as plt
import numpy as np
from treeffuser import Treeffuser
import pandas as pd
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
```

# Preprocessing
copied over from linear regression, random forest, knn model 

```python
df = pd.read_csv("final.csv")
df["EVENT_START_DATETIME"] = pd.to_datetime(df["EVENT_START_DATETIME"])
event_start = pd.to_datetime(df["EVENT_START_DATETIME"]).astype(np.int64) // 10**9
scrape_time = pd.to_datetime(df["scrape_time"]).astype(np.int64) // 10**9
df["time_to_concert"] =  event_start - scrape_time

localized_times = df.apply(lambda row: row['EVENT_START_DATETIME'].tz_convert(row['VENUE_TIMEZONE']), axis=1)

def get_day_of_week(dt):
    return dt.weekday()
df["day_of_week"] = localized_times.map(get_day_of_week)

def get_hour(dt):
    return dt.hour
df["hour_of_day"] = localized_times.map(get_hour)
to_keep = [
    "CLASSIFICATION_GENRE",
    "CLASSIFICATION_SUB_GENRE",
    "MIN_PRICE", 
    "MAX_PRICE",
    "HOT_EVENT",
    "CAPACITY",
    "population",
    "time_to_concert",
    "artist_popularity",
    "TRANSACTABLE",
    "day_of_week",
    "hour_of_day"
]
df = df[to_keep]
df["CLASSIFICATION_SUB_GENRE"] = df['CLASSIFICATION_SUB_GENRE'].fillna(df['CLASSIFICATION_GENRE'])
df.dropna(inplace=True)
categorical_cols = [col_name for col_name in df.columns if df[col_name].dtype == "object"]
X = pd.get_dummies(df.drop(["MIN_PRICE", "MAX_PRICE"], axis=1), columns=categorical_cols).to_numpy()
min_y = df['MIN_PRICE'].to_numpy()
max_y = df["MAX_PRICE"].to_numpy()
```

regarding model.fit: even though docs: https://blei-lab.github.io/treeffuser/docs/treeffuser.html say that X can be a pandas dataframe, or that there is a cat_idx parameter where the indices of categorical features can be specified, neither of these works.

```python
X_train, X_test, y_min_train, y_min_test = train_test_split(X, min_y, test_size=0.2, random_state=100)
model = Treeffuser(seed=100)
model.fit(X_train, y_min_train)
y_min_pred = model.predict(X_test)
mse = mean_squared_error(y_min_test, y_min_pred)
r2 = r2_score(y_min_test, y_min_pred)
print("mse is", mse, "and r2 is", r2)
```

```python
X_train, X_test, y_max_train, y_max_test = train_test_split(X, max_y, test_size=0.2, random_state=100)
model = Treeffuser(seed=100)
model.fit(X_train, y_max_train)
y_max_pred = model.predict(X_test)
mse = mean_squared_error(y_max_test, y_max_pred)
r2 = r2_score(y_max_test, y_max_pred)
print("mse is", mse, "and r2 is", r2)
```
