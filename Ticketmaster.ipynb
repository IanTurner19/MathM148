```python
import requests

# Replace with your API key
api_key = "ljmJ6y8jH7XpdY1iJ68nADTczTfPhVDS"

# Venue search query (e.g., searching for "Hollywood Bowl")
url = f"https://app.ticketmaster.com/discovery/v2/venues.json?keyword=Hollywood+Bowl&apikey={api_key}"

response = requests.get(url)

if response.status_code == 200:
    venues = response.json()
    if "_embedded" in venues:
        for venue in venues["_embedded"]["venues"]:
            print(f"Venue Name: {venue['name']}")
            print(f"City: {venue['city']['name']}")
            print(f"Address: {venue['address']['line1'] if 'line1' in venue['address'] else 'N/A'}")
            print(f"Postal Code: {venue['postalCode']}")
            print(f"Country: {venue['country']['name']}\n")
    else:
        print("No venues found.")
else:
    print("Error:", response.status_code, response.text)
```

```python
import requests

# Replace with your Ticketmaster API key
api_key = "ljmJ6y8jH7XpdY1iJ68nADTczTfPhVDS"

# Search query for Taylor Swift events
url = f"https://app.ticketmaster.com/discovery/v2/events.json?keyword=Taylor+Swift&classificationName=music&apikey={api_key}"

response = requests.get(url)

if response.status_code == 200:
    events = response.json()
    if "_embedded" in events:
        for event in events["_embedded"]["events"]:
            print(f"Event Name: {event['name']}")
            print(f"Date: {event['dates']['start']['localDate']}")
            print(f"Venue: {event['_embedded']['venues'][0]['name']}")
            print(f"City: {event['_embedded']['venues'][0]['city']['name']}")
            print(f"Country: {event['_embedded']['venues'][0]['country']['name']}")
            print(f"Ticket URL: {event['url']}\n")
    else:
        print("No events found for Taylor Swift.")
else:
    print("Error:", response.status_code, response.text)
```
