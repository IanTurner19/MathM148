```python
import pandas as pd
import numpy as np
```

```python
path = r"C:\Users\Nathanael Nam\Desktop\archive\uscities.csv"
cities = pd.read_csv(path)
```

```python
cities_updated = cities[['city', 'city_ascii', 'state_id', 'state_name', 'population']]
cities_updated
```

```python
path2 = r"C:\Users\Nathanael Nam\Desktop\archive\whole_quarter.csv"
whole_quarter = pd.read_csv(path2)
whole_quarter['VENUE_CITY'] = whole_quarter['VENUE_CITY'].str.strip()
whole_quarter['VENUE_CITY'] = whole_quarter['VENUE_CITY'].replace('Ft Lauderdale', 'Fort Lauderdale')
whole_quarter['VENUE_CITY'] = whole_quarter['VENUE_CITY'].replace('Hollywood', 'Los Angeles')
whole_quarter['VENUE_CITY'] = whole_quarter['VENUE_CITY'].replace('Woodside', 'New York')
whole_quarter['VENUE_CITY'] = whole_quarter['VENUE_CITY'].replace('Huntington', 'New York')
whole_quarter['VENUE_CITY'] = whole_quarter['VENUE_CITY'].replace('Wilkes Barre', 'Wilkes-Barre')
```

```python
merged_df = pd.merge(whole_quarter, cities_updated, left_on = ['VENUE_CITY','VENUE_STATE_CODE'], right_on = ['city', 'state_id'], how="left")
```
